## Machine Learning

**Creating X and y**

```python
X = pd.get_dummies(df.drop('species', axis=1), drop_first=True)
```

```python
y = df['species']
```

**Train | Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

**Decision Tree Base Model**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
base_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
print(classification_report(y_test, base_predictions))
```

```python
plot_confusion_matrix(model, X_test, y_test)
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns, data=model.feature_importances_, 
             columns=['Feature Importance']).sort_values('Feature Importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8), dpi=200)
plot_tree(model, feature_names=X.columns, filled=True);
```

**Exploring Decision Tree Hyperparameters**

```python
def report_model(model):
    model_predictions = model.predict(X_test)
    print(classification_report(y_test, model_predictions))
    print('\n')
    plt.figure(figsize=(12,8), dpi=200)
    plot_tree(model, feature_names=X.columns, filled=True);
```

```python
prunned_tree = DecisionTreeClassifier(max_depth=2)
```

```python
prunned_tree.fit(X_train, y_train)
```

```python
report_model(prunned_tree)
```

```python
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train, y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train, y_train)
```

```python
report_model(entropy_tree)
```
